```python
@dataclass
class Triangle :
    a : int|float
    b : int|float
    c : int|float
        
    def Perimeter(self):
        if (
            (self.a + self.b > self.c) and 
            (self.b + self.c > self.a)  and 
            (self.c + self.a > self.b) ):
             return self.a + self.b + self.c
             
      
    def Area(self):
        if self.Perimeter() is not None:
            s = self.Perimeter()/2
            return (s*(s-self.a)*(s-self.b)*(s-self.c))**(1/2)
        else:
            return 'Perimeter is None'
```
