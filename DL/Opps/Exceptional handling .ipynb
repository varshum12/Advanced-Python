```python
def Division2():
    try:
         num = float(input('enter int or float here'))
         den = float(input('enter int or float here'))
         div  = num/den
         print(f'Division of {num}  by  {den} is {div}')
         return div
    except Exception as e:
        print(f'Exception Occured {e}')
```

```python
def Division3():
    try:
         num = float(input('enter int or float here'))
         den = float(input('enter int or float here'))
         div  = num/den
         print(f'Division of {num}  by  {den} is {div}')
         return div
    except ValueError:
        print('Please do not insert blanck or any other type of data')
    except ZeroDivisionError:
        print('Denominator should not be 0')
    except Exception as e:
        print(f'Exception Occured {e}')
```
