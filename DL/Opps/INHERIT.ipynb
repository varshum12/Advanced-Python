## MultiLevel Inheritance

```python
class Employee:
    def __init__(self, name:str ,emp_id : int):
        self.name = name 
        self.emp_id  = emp_id

    def get_employee_info(self):
        print(f"Name is {self.name} and Employee id is {self.emp_id}")
    

class Manager(Employee):
    def __init__(self, name:str , emp_id:int , Department: str):
        super().__init__(name ,emp_id)
        self.Department  = Department

    def get_manager_info(self):
        super().get_employee_info()
        print(f'Department : {self.Department}')
    

class ProjectManager(Manager):
    def __init__(self, name , emp_id , Department , Project):
        super().__init__(name , emp_id , Department)
        self.Project =Project

    
    def get_project_manager_info(self):
        super().get_manager_info()
        print(f'Project Manager : {self.Project}')
```

### Hierarchical Inheritance

```python
from dataclasses import dataclass
```

```python
@dataclass
class User:
    name  : str
    email : str
    
    def get_user_info(self):
        print('name is --> ' , self.name)
        print('Email is -->' , self.email)

class Admin(User):
    def __init__(self, name , email , access : str):
        super().__init__(name , email)
        self.access = access 

    def get_admin_info(self):
        super().get_user_info()
        print(f"Access:{self.access}")

class RegularUser(User):
    def __init__(self, name ,email , subscription : bool):
        super().__init__(name , email)
        self.subscription = subscription

    def get_RegularUser_info(self):
        super().get_user_info()
        print(f'Subscription : {self.subscription}')
```

```python
u1  = User(name  = 'Sarthak' , email = 'Sarthak@test.com')
```

```python
u1.name
```

```python
u1.email
```

```python
A1 = Admin(name  = 'Shreesha' , email = 'Shreesha@gmail.com' , access= 'Partial')
```

```python
print(A1.name)
print(A1.email)
print(A1.access)
```

```python
A1.get_admin_info()
```

```python
RU = RegularUser(name  ='Shreya' , email= 'shreya@gmail.com' , subscription= True)
```

```python
print(RU.name)
print(RU.email)
print(RU.subscription)
```

```python
RU.get_RegularUser_info()
```
